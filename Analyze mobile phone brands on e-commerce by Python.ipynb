```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

df = pd.read_excel("D:/DATA ANALYST PORTFOLIO PROJECT/METRIC_DA/DA_Metric.xlsx")
```

```python
df.describe()
```

```python
#Change the column 'Tên sản phẩm' to lowercase
df['Tên sản phẩm'] = df['Tên sản phẩm'].str.lower()
df
```

```python
#Filter the data to exclude the irrelevant categories'Điện Gia Dụng', 'Máy Tính, Laptop & Thiết Bị Văn Phòng', 'Nhà Cửa - Đời Sống', 'Thiết Bị Số & Gaming'
desired_values = ['Điện Thoại & Máy Tính Bảng','Chưa phân loại']
fil_df = df[df['Ngành hàng'].isin(desired_values)]
fil_df
```

```python
# List the names of mobile phone brands
brand_names = {"apple", "asus", "google", "infinix", "itel", "kudixiong", "lg", "nokia", "oppo", "poco",
            "realme", "redmi", "samsung", "sony", "tecno", "vertu", "vivo", "vsmart", "xiaomi"}

# Create a new column named 'Tên thương hiệu' starting with the NaN value
fil_df['Tên thương hiệu'] = None

# If the 'Tên sản phẩm' column contains any words matching those listed in 'Tên thương hiệu' then assign that word to the 'Tên thương hiệu' column.'
for keyword in brand_names:
    mask = fil_df['Tên sản phẩm'].str.contains(keyword, case=False, regex=True)
    fil_df.loc[mask, 'Tên thương hiệu'] = keyword
fil_df
```

```python
# Check if the 'Tên thương hiệu' column has any null values.
null_values_brand = fil_df[fil_df['Tên thương hiệu'].isnull()]
null_values_brand
```

```python
#Replace null value by brand name
mask = fil_df['Tên sản phẩm'].str.contains('galaxy|ss galaxy', case=False, regex=True)
fil_df.loc[mask & fil_df['Tên thương hiệu'].isnull(), 'Tên thương hiệu'] = 'samsung'
fil_df
```

```python
# Check if the 'Tên thương hiệu' column has any null values. 
null_values = fil_df[fil_df['Tên thương hiệu'].isnull()]
null_values
```

```python
# Assign the country of origin for each brand name
conditions = [
    (fil_df['Tên thương hiệu'].isin(['apple','google'])),
    (fil_df['Tên thương hiệu'].isin(['samsung', 'lg'])),
    (fil_df['Tên thương hiệu'].isin(['oppo', 'xiaomi', 'infinix', 'itel', 'kudixiong', 'poco', 'realme', 'redmi', 'tecno', 'vivo'])),
    (fil_df['Tên thương hiệu'] == 'asus'),
    (fil_df['Tên thương hiệu'] == 'nokia'),
    (fil_df['Tên thương hiệu'] == 'sony'),
    (fil_df['Tên thương hiệu'] == 'vertu'),
    (fil_df['Tên thương hiệu'] == 'vsmart')
]

values = ['Mỹ', 'Hàn Quốc', 'Trung Quốc', 'Đài Loan', 'Phần Lan', 'Nhật Bản', 'Anh Quốc', 'Việt Nam']
                                    
# Create 'Xuất xứ' column
fil_df['Xuất xứ'] = np.select(conditions, values, default=None)
fil_df
```

```python
# Check if 'Xuất xứ' column has any null value
null_values_country = fil_df[fil_df['Xuất xứ'].isnull()]
null_values_country
```

```python
# Standardize the string structure for the 'Shop Link' column.
find_www = fil_df['Link shop'].str.contains('https://www.', case=False, regex=True)
fil_df['Link shop update'] = np.where(find_www, fil_df['Link shop'].str.replace('https://www.', 'https://'), fil_df['Link shop'])
fil_df['Link shop update']
```

```python
# Create 'Tên sàn điện tử' column
fil_df['Tên sàn điện tử'] = fil_df['Link shop update'].str.extract(r'https://(.+?)\.')
fil_df
```

```python
# Filter out non-duplicate values in the 'Tên sàn điện tử' column.
unique_values = fil_df['Tên sàn điện tử'][~fil_df['Tên sàn điện tử'].duplicated()]
unique_values
```

```python
#Classify Samsung into Samsung Galaxy S23 and other models.
fil_df['Phân loại ss'] = fil_df.apply(lambda row: 'samsung galaxy s23' if (row['Tên thương hiệu'] == 'samsung' and 's23' in row['Tên sản phẩm'])
                              else 'samsung-others' if row['Tên thương hiệu'] == 'samsung'
                              else row['Tên thương hiệu'], axis=1)
fil_df
```

```python
#Count the number of each value in the 'Phân loại ss' column
value_counts = fil_df['Phân loại ss'].value_counts()
value_counts
```

```python
# Calculate the total revenue and total quantity by brand name.
brand_stats = fil_df.groupby('Tên thương hiệu').agg({'Doanh thu': 'sum', 'Số đã bán': 'sum'}).reset_index()
brand_stats
```

```python
# Visualize the total revenue and total quantity by brand name. 
##Create figure
fig, ax1 = plt.subplots(figsize=(11, 6))

##Visualize total quantity in bar chart
sns.barplot(x='Tên thương hiệu', y='Số đã bán', data=brand_stats, ax=ax1, color='teal', errorbar=None)
ax1.set_ylabel('Số lượng đã bán', fontsize = 8)
ax1.set_xlabel('Tên thương hiệu', fontsize = 8)
ax1.set_title('Số lượng đã bán & Doanh thu theo thương hiệu', fontsize= 13)
ax1.tick_params(axis='x', labelrotation=45, labelsize=8)
ax1.tick_params(axis='y',  labelsize=8)

## Create the sencondary y-axis
ax2 = ax1.twinx()

## Visualize total revenue in bar chart
sns.lineplot(x='Tên thương hiệu', y='Doanh thu', data=brand_stats, ax=ax2, color='orange', err_style=None, linewidth=3)
ax2.set_ylabel('Doanh thu', fontsize=8)
ax2.tick_params(axis='y', labelsize=8)

plt.show()
```

```python
# Calculate total quantity by country of origin
origin_qty_stats = fil_df.groupby('Xuất xứ').agg({'Số đã bán': 'sum'}).reset_index()
# Percentage of total quantity by country of origin 
origin_qty_stats['Tỷ lệ số lượng đã bán (%)'] = (origin_qty_stats['Số đã bán'] / origin_qty_stats['Số đã bán'].sum()) * 100
# Sort in descending order by total quantity
origin_qty_stats = origin_qty_stats.sort_values(by='Tỷ lệ số lượng đã bán (%)', ascending=False)
origin_qty_stats
```

```python
# Calculate total revenue by country of origin
origin_revenue_stats = fil_df.groupby('Xuất xứ').agg({'Doanh thu': 'sum'}).reset_index()
# Percentage of total revenue by country of origin 
origin_revenue_stats['Tỷ lệ doanh thu (%)'] = (origin_revenue_stats['Doanh thu'] / origin_revenue_stats['Doanh thu'].sum()) * 100
# Sort in descending order by total revenue
origin_revenue_stats = origin_revenue_stats.sort_values(by='Tỷ lệ doanh thu (%)', ascending=False)
origin_revenue_stats
```

```python
# Visualize total revenue, total quantity, % total revenue and % total quantity by contry of origin
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(16, 18))

## Visualize total quantity by country of origin in bar chart 
sns.barplot(x='Xuất xứ', y='Số đã bán', data=origin_qty_stats, ax=ax1, color='orange', errorbar=None)
ax1.set_ylabel('Số lượng đã bán', fontsize = 8)
ax1.set_title('Số lượng đã bán theo xuất xứ', fontsize= 14)
ax1.tick_params(axis='x', labelrotation=45, labelsize=8)
ax1.tick_params(axis='y',  labelsize=8)

## Visualize total revenue by country of origin in bar chart 
sns.barplot(x='Xuất xứ', y='Doanh thu', data=origin_revenue_stats, ax=ax2, color='steelblue', errorbar=None)
ax2.set_ylabel('Doanh thu', fontsize = 8)
ax2.set_title('Doanh thu theo xuất xứ', fontsize= 14)
ax2.tick_params(axis='x', labelrotation=45, labelsize=8)
ax2.tick_params(axis='y',  labelsize=8)

## Visualize percentage of total quantity by country of origin in pie chart
ax3.pie(origin_qty_stats['Tỷ lệ số lượng đã bán (%)'], 
        labels=origin_qty_stats['Xuất xứ'], 
        autopct='%1.1f%%', startangle=90, 
        colors=sns.color_palette('flare'))
ax3.set_title('Tỷ lệ phần trăm số lượng đã bán theo Xuất xứ', fontsize= 14)

## Visualize percentage of total revenue by country of origin in pie chart
ax4.pie(origin_revenue_stats['Tỷ lệ doanh thu (%)'], 
        labels=origin_revenue_stats['Xuất xứ'], 
        autopct='%1.1f%%', startangle=90, 
        colors=sns.color_palette('ch:s=.25,rot=-.25'))
ax4.set_title('Tỷ lệ phần trăm doanh thu theo Xuất xứ', fontsize= 14)
```

```python
# Percentage of sales revenue, quantity sold of Samsung Galaxy S23 compared to other Samsung phones on Shopee platform

## Filt date by Shopee, Samsung in 'Tên sàn điện tử' & 'Tên thương hiệu' column
shopee_samsung_data = fil_df[(fil_df['Tên sàn điện tử'] == 'shopee') & (fil_df['Tên thương hiệu'] == 'samsung')]

# Calculate revenue and quantity of Samsung Galaxy S23 & Samsung-others
s23 = shopee_samsung_data[shopee_samsung_data['Phân loại ss'] == 'samsung galaxy s23']
others = shopee_samsung_data[shopee_samsung_data['Phân loại ss'] == 'samsung-others']
total_sales = shopee_samsung_data['Doanh thu'].sum()
total_qty = shopee_samsung_data['Số đã bán'].sum()

# Calculate percentage of revenue and quantity of Samsung Galaxy S23 & Samsung-others
s23_sales_pct = (s23['Doanh thu'].sum() / total_sales) * 100
other_sales_pct = 100 - s23_sales_pct
s23_qty_pct = (s23['Số đã bán'].sum() / total_qty) * 100
other_qty_pct = 100 - s23_qty_pct

print(f'Tỷ lệ doanh số của samsung galaxy s23 trên shopee: {s23_sales_pct:.2f}%')
print(f'Tỷ lệ doanh số của samsung-others trên shopee: {other_sales_pct:.2f}%')
print(f'Tỷ lệ số đã bán của samsung galaxy s23 trên shopee: {s23_qty_pct:.2f}%')
print(f'Tỷ lệ số đã bán của samsung-others trên shopee: {other_qty_pct:.2f}%')
```

```python
# Visialize percentage of sales revenue, quantity sold of Samsung Galaxy S23 compared to other Samsung phones on Shopee platform
## Create dataframe 
data = {'labels': ['SS Galaxy S23', 'SS-others'],
    'sizes_sales': [s23_sales_pct, other_sales_pct],
    'sizes_qty': [s23_qty_pct, other_qty_pct]
        }
df_pie = pd.DataFrame(data)

## Visualize % quantity sold & revenue of Samsung Galaxy S23 compared to other Samsung phones on Shopee platform in pie chart
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax1.pie(df_pie['sizes_qty'], 
        labels=df_pie['labels'], 
        autopct='%1.1f%%', startangle=90, 
        textprops={'fontsize': 8},
        colors=sns.color_palette('blend:#7AB,#EDA'))
ax1.set_title('Tỷ lệ số đã bán của SS Galaxy S23 trên Shopee', fontsize= 12)

ax2.pie(df_pie['sizes_sales'], 
        labels=df_pie['labels'], 
        autopct='%1.1f%%', startangle=90, 
        textprops={'fontsize': 8},
        colors=sns.color_palette('dark:#5A9_r'))
ax2.set_title('Tỷ lệ doanh thu của SS Galaxy S23 trên Shopee', fontsize= 12)

plt.show()
```
